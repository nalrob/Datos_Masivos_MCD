
# Summary by Albums 

```python
import pandas as pd
import plotly.express as px

# Obtener los datos de los álbumes de Spotify
album_name = []
artist_name = []
release_date = []
total_tracks = []

# Rango de años
for year in range(1990, 2021):
    album_results = sp.search(q=f'year:{year}', type='album', limit=50)
    
    # Itera sobre los resultados de álbumes
    for album in album_results['albums']['items']:
        album_name.append(album['name'])
        artist_name.append(album['artists'][0]['name'])
        release_date.append(album['release_date'])
        total_tracks.append(album['total_tracks'])

# Crea el DataFrame a partir de las listas de datos
album_dataframe = pd.DataFrame({'album_name': album_name, 'artist_name': artist_name, 'release_date': release_date, 'total_tracks': total_tracks})

print(album_dataframe.shape)
album_dataframe.head(10)
```

```python
# Extraer el año de la fecha de lanzamiento
album_dataframe['Year'] = pd.to_datetime(album_dataframe['release_date']).dt.year

# Calcular la media de canciones por año
mean_tracks_by_year = album_dataframe.groupby('Year')['total_tracks'].mean().reset_index()

# Crear el gráfico de caja y línea de tendencia utilizando Plotly
fig = px.scatter(mean_tracks_by_year, x='Year', y='total_tracks', title='Media de canciones por año',
                 trendline='ols')
fig.show()
```

Esta información, no es muy reelevante, pero se observa que la media de canciones por album año a año va a la baja. 

Al parecer cada vez los artistas graban menos canciones por album.
